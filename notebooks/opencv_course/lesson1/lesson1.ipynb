```python
img_path = os.path.join(os.getcwd() + "/images/opencv_bootcamp_assets_NB1.zip")
```

```python
def download_and_extract(url, save_path):
    # file_name=urlretrieve(url,path)
    urlretrieve(url, save_path)
    file_name = save_path.split("/")[-1]
    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(file_name) as z:
            print(z)
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
```

```python
asset_zip_path = os.path.join(os.getcwd(), "images/opencv_bootcamp_assets_NB1.zip")
if not os.path.exists(asset_zip_path):
    download_and_extract(URL, asset_zip_path)
```
