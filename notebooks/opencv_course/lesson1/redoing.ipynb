```python
new_zealand = "images/New_Zealand_Lake.jpg"
nz_bgr = cv2.imread("images/New_Zealand_Lake.jpg", cv2.IMREAD_COLOR)
```

```python
plt.imshow(nz_bgr)
```

```python
cv2.IMREAD_COLOR
```

```python
nz_bgr.shape
```

```python
Image(new_zealand)
```

```python
b, g, r = cv2.split(nz_bgr)
```

```python
plt.figure(figsize=[20, 5])
plt.subplot(141)
plt.imshow(r, cmap="gray")
plt.title("Red Channel")
plt.subplot(142)
plt.imshow(g, cmap="gray")
plt.title("Green Channel")
plt.subplot(143)
plt.imshow(b, cmap="gray")
plt.title("Blue Channel")
new_zealand_merged = cv2.merge((b, g, r))
plt.subplot(144)
plt.imshow(new_zealand_merged[:, :, ::-1])
plt.title("Merged Output")
```

```python
nz_rgb = cv2.cvtColor(nz_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(nz_rgb)
```
